```python
# 1) 배치 학습(batch learning)
# 모든 학습 표본을 한 번에 사용해서 업데이트 양을 구하는 배치 학습
# 모든 학습 표본으로부터 구한 업데이트 양으로 신경망을 업데이트하고, 이렇게 수정된 신경망을 이용하여 모든 학습을 평가한다.
#  그리고 다시 업데이트 양을 구하고 신경망을 수정하는 과정을 반복한다.

# 배치 학습: 모든 학습 표본으로부터 업데이트 양을 계산하므로, 학습 집합에 포함될 수 있는 잡음 데이터의 영향을 억제할 수 있다.
# 그러나 한 번 데이터를 하는 데 학습 표본 모두를 사용하기 때문에, 계산하는 데 많은 시간이 걸린다.
```

```python
# 2) 순차적 학습: sequential learning or online learning
# 학습 표본을 하나씩 노출시키는 순차적 학습.

# 표본 하나로 신경망을 수정한 후, 다음 표본을 수정된 신경망으로 평가한다.
# 그리고 그 표본으로 다시 업데이트 양을 구해서 신경망을 수정한다.

# 순차적 학습은 표본 하나를 가지고 업데이트 양을 결정하므로 표본에 따라 업데이트 양이 크게 변화할 수 있다.
# 그 때문에 학습이 수렴하지 않는 경우가 있다.
```

```python
# 3) 미니 배치 학습(mini-batch learning)
# 미니 배치 학습: 학습 집합을 몇 개의 부분 집합 D로 나누어서 그 부분집합 하나로부터 1회 수정을 수행한다.
# 모든 부분집합에 대한 수정이 끝나면, 다시 첫 부분 집합으로 돌아가 반복한다.

# 미니배치학습은 한 번에 모든 학습 표본을 사용하지 않고, 적은 수의 표본으로 구성된 부분 집합을 사용하므로,
# 한 번에 업데이트를 하는 데, 배치 학습보다 시간이 덜 걸린다.

# 또, 여러 개의 학습 표본으로 업데이트 양을 결정하므로, 순차적 학습처럼 업데이트 양이 크게 변동하는 일도 없다.
# 이렇듯 미니배치 학습은 순차적 학습과 배치 학습의 장점만을 모은 방법이라고 할 수 있다.

# 확률적 경사 하강법: SGD(Stochastic Gradient Descent)
```
