**합성곱 신경망**

```python
# CNN: Convolutional Neural Network

# 합성곱 신경망: 입력층, 합성곱층(convolution layer), 풀링층(pooling layer), 전결합층(fully-connected layer), 출력층(output layer)으로 구성된다.
# 합성곱층과 풀링층은 여러 층 번갈아 가며, 배치되어 깊은 신경망을 형성하고, 그 뒤에 전결합층이 다시 몇 층 이어진다.
```

**합성곱층**

```python
# 출처: https://wikidocs.net/62306
# 합성곱 처리: 입력에 대한 가중 필터의 내적을 계산하면서 수행된다.

# 합성곱층: 입력 표본에 대하여 합성곱 처리를 수행하고, 특징 맵(feature map)을 얻는다.
# 합성곱층은 같은 가중 필터로 모든 입력 표본에 합성곱 처리를 수행한다.

# 두 번째 이후의 합성곱층은 이전 층의 특징 맵을 입력으로 받아 합성곱 처리를 수행한다.
# 합성곱 연산의 결과값을 그대로 특징 맵의 값으로 하지 않고, 활성화 함수를 거친 출력을 특징 맵의 값으로 한다.
# 합성곱이 받는 입력이 3채널의 컬러 이미지일 경우, 가중필터는 3xMxM처럼 3차원이 된다. (M: 필터의 크기)

# 특징 맵의 수는 바로 이전 합성곱 층의 가중 필터 수에 따라 변한다.
# 예를 들어, 바로 이전 합성곱층의 가중 필터 수가 8이었다면, 다음 합성곱층의 필터는 8 x M x M, 3차원 가중 필터가 된다.
```

**풀링층**

```python
# 풀링층은 합성곱층에서 출력된 특징 맵을 축소한다.
# 이때, 주목하는 영역을 설정하고 그 영역의 특징 맵 값으로부터 새로운 특징 맵 값을 계산한다.

# 주목 영역의 크기가 2 x 2인 경우, 새로운 특징 맵은 1/4로 크기가 줄어든다.
# 이를 통해 기하학적 변화에 대한 불변성을 획득할 수 있다.

# 가장 자주 쓰이는 것이 최대 풀링이다.
# 최대 풀링은 주목 영역 내에서 가장 큰 값을 새로운 특징 맵의 값으로 한다.
```
